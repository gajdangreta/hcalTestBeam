```python
ped_1=pedestal_df[pedestal_df['end']==1].pedestal
std_dev_1=pedestal_df[pedestal_df['end']==1].std_dev
ppts_1=pedestal_df[pedestal_df['end']==1].pedestal_per_time_sample
ppts_std_dev_1=pedestal_df[pedestal_df['end']==1].pedestal_per_time_sample_std_dev
```

```python
ped_fixed=pedestal_df[pedestal_df['end']==0]
ped_fixed=ped_fixed.rename(columns={"pedestal": "pedestal_end0",
                                    "std_dev" : "std_dev_end0",
                                    "pedestal_per_time_sample":"pedestal_per_time_sample_end0",
                                    "pedestal_per_time_sample_std_dev":"pedestal_per_time_sample_std_dev_end0"})
ped_fixed=ped_fixed.assign(pedestal_end1=ped_1.values)
ped_fixed=ped_fixed.assign(std_dev_end1=std_dev_1.values)
ped_fixed=ped_fixed.assign(pedestal_per_time_sample_end1=ppts_1.values)
ped_fixed=ped_fixed.assign(pedestal_per_time_sample_std_dev_end1=ppts_std_dev_1.values)

ped_fixed.drop(["mean", "pedestal_per_time_sample_mean", "end"], axis=1, inplace=True)
ped_fixed.reset_index(inplace=True, drop=True)
```
