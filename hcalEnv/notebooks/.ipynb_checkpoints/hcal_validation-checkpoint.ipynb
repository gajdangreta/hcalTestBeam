```python
import sys
sys.path.append("../src")

from event_selections import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import norm
from tqdm import tqdm
```

```python
c_folder="../../data/calibration"
d_folder="../../data/runData"
p_folder="../../plots/energy_distributions/"
```

```python
def gaussian(x, mu, sig, A):
    return (A / (np.sqrt(2.0 * np.pi) * sig) * np.exp(-np.power((x - mu) / sig, 2.0) / 2))
```

```python
def fit_gaussian(energies):
    values, bin_edge=np.histogram(energies, bins=30)
    bins=(bin_edge[:-1] + bin_edge[1:]) / 2

    max_val = max(values)
    idx_max = list(values).index(max_val)
    mu_p0=bins[idx_max]
    
    param, pcov=curve_fit(gaussian, bins, values, p0=[mu_p0,100,6000])
    perr=np.sqrt(np.diag(pcov))
    
    return param, perr
```

```python
def linear(x,a,b):
    return a*x+b
```

```python
def make_energy_profile(run_number):
    # p_folder="../../plots/run_"+str(run_number)
    p_folder="../../plots/energy_distributions/"
    clean_run=import_and_select(c_folder,d_folder,run_number, cut_first_layer=True,cut_back_layers=True, use_sigma=True, back_layers=5)

    summed=clean_run.groupby(["event"]).sum()
    energy_array=((summed["adc_sum_end0"]+summed["adc_sum_end1"]).values)/2

    param, perr=fit_gaussian(energy_array)

    xnums=np.linspace(min(energy_array), max(energy_array),1000)

    plt.hist(energy_array, bins=30, color='magenta', alpha=0.5, edgecolor='black', zorder=1)
    plt.plot(xnums, gaussian(xnums,*param), color='black', zorder=2, 
         label=r"Gaussian fit:""\n"" $\mu$="+str(round(param[0],2))+", $\sigma$="+str(round(param[1],2)))

    plt.title(r"Run "+str(run_number))
    plt.xlabel("Average energy (MeV)")
    plt.ylabel("Frequency")
    plt.legend(loc='upper right', fontsize=10)
    
    # plt.savefig(p_folder+"/energy_dist.png", bbox_inches='tight')
    plt.savefig(p_folder+"run_"+str(run_number)+".png", bbox_inches='tight')
    plt.show()
    return param,perr
```

```python
run_list=[300,301,302,303,304,305,307]
# run_list=[305,307]
data=[]
for run in run_list:
    run_params, param_errs=make_energy_profile(run)

    line=[int(run)]
    line=np.append(line,run_params[:2])
    line=np.append(line,param_errs[:2])
    data.append(line)
```

```python
fit_data=pd.DataFrame(data,columns=["run_number","mu", "delta_mu", "sigma", "delta_sigma"])
fit_data["run_number"]=fit_data["run_number"].astype(int)
fit_data=fit_data.set_index("run_number")
```

```python
run_info=pd.read_csv(c_folder+"/runs.csv", index_col="run_number").drop("Unnamed: 0", axis=1)
```

```python
full_df=fit_data.merge(run_info, left_on="run_number", right_on="run_number")
full_df
```

```python
params, _=curve_fit(linear,full_df["beam_energy"], full_df["mu"], sigma=full_df["delta_mu"])
params2, _=curve_fit(linear,full_df["beam_energy"][1:], full_df["mu"][1:], sigma=full_df["delta_mu"][1:])

print(params)
print(params2)
```

```python
xes=np.linspace(0,5,1000)

plt.errorbar(full_df["beam_energy"],full_df["mu"],full_df["delta_mu"], linestyle='', color="darkviolet", zorder=2)
plt.scatter(full_df["beam_energy"],full_df["mu"], color="magenta", zorder=3)
plt.plot(xes,linear(xes,*params), color="lightgray", linestyle="dashed", label="Linear fit of points")
plt.plot(xes,linear(xes,*params2), color="dimgray", linestyle="-",label="Linear fit (excluding 4 GeV)")
plt.ylabel("Average reconstructed energy [MeV]")
plt.xlabel("Beam energy [GeV]")
plt.grid(True)
plt.legend()
plt.savefig(p_folder+"/reconstructed.png", bbox_inches='tight')
```
